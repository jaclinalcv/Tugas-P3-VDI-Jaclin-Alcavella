### 2. Menentukan Tujuan, User, Fungsi, dan Tone

- User : Manajemen perusahaan

- Fungsi : untuk membantu user mengetahui dan memahami kategori terlaris serta bulan dengan tingkat penjuaan tertinggi dengan mudah

### 3. Membuat Visualisasi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh data penjualan
data = {
    'Bulan': ['Okt', 'Nov', 'Des', 'Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep'],
    'Kemeja': [50, 45, 60, 15, 70, 65, 80, 75, 90, 85, 100, 95],
    'Kaos': [30, 25, 40, 35, 50, 45, 20, 55, 70, 65, 80, 50],
    'Blouse': [20, 15, 30, 25, 40, 60, 50, 45, 60, 25, 70, 65]
}

# Mengonversi dan menampilkan data menjadi DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# Membuat visualisasi menggunakan seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Bulan', y='value', hue='variable',
             data=pd.melt(df, ['Bulan']), marker="o")

# Memberi judul dan label pada visualisasi
plt.title('Visualisasi Penjualan Bulanan Berdasarkan Kategori Produk', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Volume Penjualan', fontsize=12)
plt.xticks(rotation=45)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

- Data yang digunakan berupa data bulanan penjualan pakaian dengan kategori kemeja, kaos, dan blouse dalam jangka waktu 12 bulan terakhir

- Visualisasi garis menunjukkan jumlah penjualan setiap bulan yang selalu berubah

- Library yang digunakan berupa seaborn dan matplotlib untuk plot yang mudah dibaca sesuai data yang dimiliki
